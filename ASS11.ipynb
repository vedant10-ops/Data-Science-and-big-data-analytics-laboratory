```python
import pandas as pd
from scipy import stats
```

```python
# Step 1: Load the dataset into a DataFrame
df = pd.read_csv('demo.csv')
```

```python
z_scores = stats.zscore(df)
threshold = 3
outlier_indices = (abs(z_scores) > threshold).any(axis=1)
```

```python
outlier_indices 
```

```python
# Print indices of outlier rows
print("Indices of outlier rows:")
print(df[outlier_indices].index)
```

```python
df
```

```python
# Step 3: Replace outliers with median values
median_values = df.median()
df[outlier_indices] = median_values
```

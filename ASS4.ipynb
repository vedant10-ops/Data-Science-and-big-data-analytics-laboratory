```python
import pandas as pd
#import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
csv_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
col_names = ['sepal_ln','sepal_wd','petal_ln','petal_wd','species']
```

```python
iris_data = pd.read_csv(csv_url, names=col_names)
```

```python
iris_data
```

```python
iris_data['species'].unique()
```

```python
onehot_encoder = OneHotEncoder()
```

```python
species_encoded = onehot_encoder.fit_transform(iris_data[['species']])
```

```python
print(species_encoded)
```

```python
species_encoded_df = pd.DataFrame(species_encoded.toarray(), columns=onehot_encoder.categories_[0])
```

```python
species_encoded_df
```

```python
iris_data_encoded = pd.concat([iris_data, species_encoded_df], axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the demo dataset into a DataFrame
# Replace 'demo.csv' with the path to your dataset
df = pd.read_csv('demo.csv')
df

df.plot(kind='box',figsize=(10,6))
plt.ylim(0,250)
```

```python
# Step 2: Calculate the IQR for each numerical column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
# Step 3: Determine lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Step 4: Identify rows with values outside the bounds
outliers = ((df < lower_bound) | (df > upper_bound)).any(axis=1)
outliers
```

```python
# Step 5: Remove rows identified as outliers
df_no_outliers = df[~outliers]
```

```python
# Display the cleaned DataFrame
print("DataFrame after removing outliers:")
print(df_no_outliers)
```

```python
df_no_outliers.plot(kind='box',figsize=(10,6))
plt.ylim(0,250)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('AcademicPerformance.csv')
df
```

```python
# Step 2: Check null values in the dataset
print("Null values in the dataset:")
print(df.isnull().sum())
```

```python
# Step 3: Replace null values with standard null value NaN
df.replace('', pd.NA, inplace=True)
```

```python
# Step 4: Dropping rows with at least 1 null value
df_dropped_rows = df.dropna()
df_dropped_rows
```

```python
# Step 5: Dropping rows if all values in that row are missing
df_dropped_all_rows = df.dropna(how='all')
df_dropped_all_rows
```

```python
# Step 6: Dropping columns with at least 1 null value
df_dropped_cols = df.dropna(axis=1)
df_dropped_cols
```

```python
# Step 7: Dropping Rows with at least 1 null value in CSV file
df.to_csv('AcademicPerformance_clean.csv', index=False)
```

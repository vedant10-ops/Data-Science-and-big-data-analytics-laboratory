```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
df = pd.read_csv('demo.csv')
```

```python
# Step 2: Detect the outliers using BoxPlot
plt.figure(figsize=(12, 6))
df.boxplot()
plt.ylim(0,500)
plt.title('Boxplot of the dataset')
plt.show()
```

```python
# Step 3: Handle the outliers using Quantile-based flooring and capping
Q1 = df.quantile(0.10)
Q3 = df.quantile(0.90)
IQR = Q3 - Q1
```

```python
# Flooring and capping the outliers
floor_value = Q1 - 1.5 * IQR
cap_value = Q3 + 1.5 * IQR
```

```python
# Applying Quantile-based flooring and capping
df_floored_capped = df.clip(lower=floor_value, upper=cap_value, axis=1)
```

```python
# Visualize the BoxPlot after handling outliers
plt.figure(figsize=(8, 6))
df_floored_capped.boxplot()
plt.ylim(0,500)

plt.title('Boxplot after handling outliers')
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
# Step 1: Load the dataset
df = pd.read_csv('AcademicPerformance.csv')
```

```python
# Step 2: Check null values in the dataset
print("Null values in the dataset:")
print(df.isnull().sum())
```

```python
# Step 3: Replace missing values with NaN
df.replace('', np.nan, inplace=True)
```

```python
df
```

```python
df['math score'] = pd.to_numeric(df['math score'], errors='coerce')
mean_math_score = df['math score'].mean()
df['math score'].fillna(mean_math_score, inplace=True)
```

```python
df
```

```python
# Step 5: Replace missing values of Reading Score with standard deviation
df['reading score'] = pd.to_numeric(df['reading score'], errors='coerce')
std_reading_score = df['reading score'].std()
df['reading score'].fillna(std_reading_score, inplace=True)
```

```python
df
```

```python
# Step 6: Replace missing values of place with common value "Nashik"
df['Region'].fillna('Nashik', inplace=True)
```

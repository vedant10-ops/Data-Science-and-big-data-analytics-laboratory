```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")
```

```python
# 1. Use Seaborn to explore patterns in the Titanic dataset
# Pairplot to visualize relationships between numeric variables
sns.pairplot(titanic, hue='survived', diag_kind='kde')
plt.suptitle("Pairplot of Titanic Dataset", y=1.02)
plt.show()
```

```python
# 2. Plot a histogram to check the distribution of ticket prices (fare) for each passenger
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='fare', bins=30, kde=True)
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
